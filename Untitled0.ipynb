```python
# -- coding: utf-8 --
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

def predict(x):
    return slope * x + intercept

# Load the dataset
df = pd.read_csv("univariate_linear_regression_dataset.csv")

# Extracting X and Y values from the dataset
X = df.iloc[:, 0]  # Selecting the first column of the dataset
Y = df.iloc[:, 1]

# Plotting the data points
plt.scatter(X, Y)
plt.grid()

# Calculating the linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

# Generating the line of best fit
fitLine = predict(X)

# Plotting the line of best fit
plt.plot(X, fitLine, c='r')

# Displaying the plot
plt.show()

# Predicting for a given value
prediction = predict(20.27)
print("Prediction for x = 20.27:", prediction)
```
